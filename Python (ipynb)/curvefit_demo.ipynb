```python
from scipy import optimize
```

```python
t,y = np.loadtxt('lab1data2.csv',delimiter=',',usecols=(0,1),unpack=True)
plt.plot(t,y,'b+')
```

```python
def fitfun(x,A,w,ph,d):
    return A*np.cos(w*x+ph)+d
```

```python
popt = optimize.curve_fit(fitfun,t,y,p0=[4,1.5,0.5,0])  # can handle nonlinear fit functions
print(popt)
```

```python
plt.plot(t,y,'b+',t,fitfun(t,*popt[0]),'r')
```
