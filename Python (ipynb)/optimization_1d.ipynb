```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(t):
    return 3*np.cos(t)-t
```

```python
def fprime(t):
    return -3*np.sin(t)-1
```

```python
stepscale = 0.1
maxmin = -1
```

```python
t = 1
n = 0
```

```python
while np.abs(fprime(t)) > 0.01 and n < 1000:
    t = t + maxmin*stepscale*fprime(t)
    n = n + 1

if n == 1000:
    print('Inconclusive.')
else:
    print(t, fun(t))
```

```python
tt = np.linspace(-10,10,1000)
```

```python
plt.plot(tt,fun(tt),'b',t,fun(t),'r*')
```
