```python
import numpy as np
import matplotlib.pyplot as plt
from random import uniform
```

```python
N = 1000
levels = 8
p = 0.5
```

```python
# initialize at 0
slots = np.zeros(levels+1)

for puck in range(N):
    loc = 0  # center of board
    
    for level in range(levels):
        x = uniform(0,1)
        
        if x < p:  # has probability p of occurring, go left
            loc = loc - 1
        else:  # probability (1-p), go right
            loc = loc + 1
    
    # interpret the location (will be one of the following: -8, -6, ..., 6, 8)
    pos = int(loc/2)  # (now -4,-3,...,3,4 and we make sure it is an integer)
    slot = pos + 4  # shift so that these are the indices of the slots variable (0,1,...,7,8)
    
    slots[slot] = slots[slot] + 1  # add 1 to the count of pucks in this slot

print(slots)  # for the number of pucks per slot
print(slots*100/N)  # for the percentage of pucks per slot
```

```python
plt.plot(range(levels+1),slots,'b')
```
