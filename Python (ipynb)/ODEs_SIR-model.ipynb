```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def sir(t,y,b,g,N):
    return [-(b/N)*y[0]*y[1], (b/N)*y[0]*y[1]-g*y[1], g*y[1]]
```

```python
tspan = [0, 60]
y0 = [1000, 1, 0]
b = 1
g = 1/14
N = sum(y0)
print('R0 = '+str(b/g))
```

```python
sol = solve_ivp(sir,tspan,y0,args=[b,g,N],max_step=0.1)
```

```python
plt.plot(sol.t,sol.y[0].T,'g',sol.t,sol.y[1].T,'r',sol.t,sol.y[2].T,'b')
```
